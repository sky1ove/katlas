## Combine pspa and ks

### Lower case

```python
pspa=Data.get_pspa().iloc[:,:-3] # remove the redundant 0 position
```

```python
ks = Data.get_ks() # normal case with lower case
ks = ks[pspa.columns]
```

```python
ks
```

List the overlap kinase and choose one of them

```python
overlap = ks.index[ks.index.isin(pspa.index)]
```

```python
# overlap.to_series().to_csv('compare.csv')
```

After choosing the one to use for combine, read back

```python
choose = pd.read_csv('raw/compare3.csv')
```

```python
choose.head()
```

```python
ks_only = ks.index[~ks.index.isin(pspa.index)]

KS_list = choose.query('choose == "KS"').kinase.tolist() + ks_only.tolist()
```

```python
len(KS_list)
```

```python
pspa_only = pspa.index[~pspa.index.isin(ks.index)]

PSPA_list = choose.query('choose == "PSPA"').kinase.tolist() + pspa_only.tolist()
```

```python
len(PSPA_list)
```

```python
ks2 = ks.loc[KS_list]

ks2['source'] = 'KS'
```

```python
pspa2 = pspa.loc[PSPA_list]

pspa2['source'] = 'PSPA'
```

```python
combine = pd.concat([ks2,pspa2],axis=0)
```

```python
combine_info = combine[['source']]
```

```python
combine = combine.iloc[:,:-1]
```

```python
# combine.to_parquet('combine_main.parquet')
combine.to_csv('raw/combine_freq.csv')
```
